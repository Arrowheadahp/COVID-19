```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
%matplotlib inline
pd.set_option('display.max_rows', 35)
plt.style.use('dark_background')
```

```python
url_country = 'https://github.com/datasets/covid-19/raw/master/data/countries-aggregated.csv'
country_df = pd.read_csv(url_country)
```

```python
country_df['Date'] = pd.to_datetime(country_df.Date)
country_df['Removed'] = country_df.Recovered + country_df.Deaths
country_df['Active'] = country_df.Confirmed - country_df.Removed
country_df['Mortality'] = country_df.Deaths/country_df.Removed
country_df['Part_Active'] = country_df.Active/country_df.Confirmed
```

```python
country_df.tail(12)
```

```python
last_day = dt.date.today() - dt.timedelta(days=1)
Last_day_df = country_df[country_df.Date == last_day].set_index('Country')
Last_day_df = Last_day_df[Last_day_df.Confirmed>2500]
```

```python
Last_day_df.sort_values('Part_Active', ascending = False)
```

```python
C = ['Confirmed', 'Active', 'Removed', 'Recovered', 'Deaths']   
```

```python
Last_day_df.sort_values('Mortality', ascending = True)
```

```python
len(Last_day_df)
```

```python
Important_countries = ['India',
                      'US', 'Italy', 'Spain',
                      'UK', 'Ireland', 'Norway',
                      'China', 'Korea, South', 'Iran'][::-1]
```

```python
Countries_Dict_DF = {country: country_df[country_df.Country == country].set_index('Date')
                        for country in Important_countries}
```

```python
Countries_Dict_DF['India'].tail(10)
```

```python
x = 'Confirmed'
for col in Last_day_df.columns[1:]:
    if col!=x:
        plt.figure(figsize = (17, 8))
        plt.title(f'{col} with respect to {x}')
        plt.xlabel(x)
        plt.ylabel(col)
        plt.xscale('log')

        if col not in ['Mortality', 'Part_Active']:
            plt.yscale('log')
        for cun in Countries_Dict_DF:
            df = Countries_Dict_DF[cun]
            plt.plot(df[x], 
                    df[col],
                    label = cun,
                    linewidth = 4
                    )
        plt.legend(Important_countries)
        plt.plot()
```

```python
for col in Last_day_df.columns[1:]:
    plt.figure(figsize = (17, 8))
    plt.title(f'{col} with respect to Date')
    plt.xlabel('Date')
    plt.ylabel(col)
    
    if col not in ['Mortality', 'Part_Active']:
        plt.yscale('log')
    for cun in Countries_Dict_DF:
        df = Countries_Dict_DF[cun]
        plt.plot(df.index, 
                df[col],
                label = cun,
                linewidth = 3
                )
    plt.legend(Important_countries)
    plt.plot()
```
